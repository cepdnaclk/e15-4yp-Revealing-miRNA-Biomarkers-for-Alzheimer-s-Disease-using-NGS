```python
# Function for get accuracy using linear SVM model
def linearSVM(dataFrame, target):

  classifer = svm.SVC(kernel='linear') 
  scores = cross_val_score(classifer, dataFrame, target, cv=3)
  avg_score = scores.mean()

  return avg_score

# Function for get accuracy using Logistic Regression model

def logisticRegression(dataFrame, target):

  classifer = LogisticRegression(random_state=42)
  scores = cross_val_score(classifer, dataFrame, target, cv=3)
  avg_score = scores.mean()

  return avg_score

# Function for get accuracy using Random Forest model

def randomForest(dataFrame, target):

  classifier = RandomForestClassifier(n_estimators=100, max_depth=2,random_state=42)
  scores = cross_val_score(classifier, dataFrame, target, cv=3)
  avg_score = scores.mean()

  return avg_score

# Function for return new Data Frame selected using correlation coefficients

def returnNewDF(dataFrame, corr, ran):
  columns = np.full((corr.shape[0],), True, dtype=bool)
  for i in range(corr.shape[0]):
      for j in range(i+1, corr.shape[0]):
          if corr.iloc[i,j] >= ran or corr.iloc[i,j] <= -ran:
              if columns[j]:
                  columns[j] = False

  new_columns = dataFrame.columns[columns]
  new_df = dataFrame[new_columns]

  return new_df
```

```python
# Function for return accuracy score for ML models linear SVM, LR and RF 
# Return the no of features according to the correlation coefficient

def returnAccuracies(dataFrame):
    
    i = 1
    list_i = []
    list_svm = []
    list_lr = []
    list_rf= []
    list_feacture_count = []

    while i >= 0.4:
        tmp_DF = dataFrame
        tmp_DF = returnNewDF(tmp_DF, tmp_DF.corr(), i)
        list_i.append(i)
        list_svm.append(linearSVM(tmp_DF, y))
        list_lr.append(logisticRegression(tmp_DF, y))
        list_rf.append(randomForest(tmp_DF , y))
        list_feacture_count.append(len(tmp_DF .columns))
        i = i - 0.0025

    rows = ["i", "linearSVM", "logisticRegression", "randomForest", "No of features"]
    data = np.array([list_i, list_svm, list_lr,list_rf, list_feacture_count])
    accuracy_DF = pd.DataFrame(data=data, index=rows).transpose()

    return accuracy_DF
```

```python
# Function to get selected features

def getFeatures(dataFrame):
  tmp_DF = pd.DataFrame( dataFrame.columns)
  return tmp_DF

# Funtion for get classification accuracies for linear SVM, LR and RF

def classificationResults(dataFrame, y):
    X_train, X_test, y_train, y_test = train_test_split(dataFrame, y, test_size=0.3, random_state=42)
    
    svm_classifier = svm.SVC(kernel='linear') 
    crossScore_svm = cross_val_score(svm_classifier, X_train, y_train, cv=3)
    svm_classifier.fit(X_train, y_train)
    svm_predicted = svm_classifier.predict(X_test)

    LR_classifier = LogisticRegression(random_state=42)
    crossScore_LR = cross_val_score(LR_classifier, X_train, y_train, cv=3)
    LR_classifier.fit(X_train, y_train)
    LR_predicted = LR_classifier.predict(X_test)
  
    RF_classifier = RandomForestClassifier(n_estimators=100, max_depth=2,random_state=42)
    crossScore_RF = cross_val_score(RF_classifier, X_train, y_train, cv=3)
    RF_classifier.fit(X_train, y_train)
    RF_predicted = RF_classifier.predict(X_test)

    print("Training and Testing accuracies of linear SVM model")
    print("Testing: " , round(metrics.accuracy_score(np.int64(y_test.values), svm_predicted)*100,2) ,"Training: ", round(crossScore_svm.mean()*100,2))
  
    print("Training and Testing accuracies of Random forest model")
    print("Testing: " , round(metrics.accuracy_score(np.int64(y_test.values), LR_predicted)*100,2) ,"Training: ", round(crossScore_LR.mean()*100,2))

    print("Training and Testing accuracies of Logistic Regression model")
    print("Testing: " , round(metrics.accuracy_score(np.int64(y_test.values), RF_predicted)*100,2) ,"Training: ", round(crossScore_RF.mean()*100,2))

```
